### Step 6. Generate subplots to display the histograms for acceleration, displacement and weight and kilometer_per_litre. Note down your insight for the same

```python
plt.subplot(2,2,1)
plt.title('acceleration')
plt.hist(mpg_df.acceleration,  color = 'b')

plt.subplot(2,2,2)
plt.title('displacement')
plt.hist(mpg_df.displacement,  color ='r')

plt.subplot(2,2,3)
plt.title('weight')
plt.hist(mpg_df.weight,  color = 'g')

plt.subplot(2,2,4)
plt.title('kilometer_per_litre')
plt.hist(mpg_df.mpg,  color = 'y')





```

### Step 7. Create a plot to check the relationship between displacement and weight for origin 2, color the datapoints based on no of cylinders. Note down your insight for the same and save the plot as an image file. 

```python
# disp vs weight for origin 2 
# hue = no of cylinders

new_df = mpg_df[mpg_df.origin == 2]
sns.scatterplot(x = 'displacement', y = 'weight', hue = 'cylinders', data = new_df)
```

### Step 8. Display the frequency distribution of the Origin variable. Note down your insight for the same

```python
sns.countplot(x = 'origin', data = mpg_df)
```

### Step 9. Check the relationship of multiple variables wrt kilometer_per_litre. Note down your insight for the same

```python
sns.pairplot(x_vars='mpg', y_vars=['cylinders','displacement','weight','acceleration','model year', 'origin'], data = mpg_df, kind ='reg')
```

### Weight vs mpg

As weight of the vehicle increases, miles per gallon decreases
this implies heavy weight vehicles get lesser miles per gallon

### Displacement vs mpg

Displacement decreases as mpg increases 
Vice versa, more displacement means lesser mpg

### Model year vs mpg
There is a noticeable increase of mpg seen in recent model cars


### Cylinder vs mpg
When more cylinders are present, miles per gallon is lesser. As number of cylinders decrease the mpg increases. 
This trend is observed for cylinders on 4 and above cylinders.
